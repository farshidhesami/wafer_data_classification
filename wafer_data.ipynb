```python
data_frames = []

# Read CSV data and handle missing values
for file_name in csv_files:
    data = pd.read_csv(file_name)

# Create a separate copy of the DataFrame to preserve the original data
cleaned_data = data.copy()

# Handle missing values (e.g., replace NaN with a specific value or perform imputation)
cleaned_data = cleaned_data.fillna(value=0)  # Replace NaN with 0

cleaned_data['File Name'] = file_name  # Add file name as a column
data_frames.append(cleaned_data)

# Concatenate all DataFrames into a single DataFrame
combined_data = pd.concat(data_frames)

# Create the "wafer_data" folder if it doesn't exist
folder_path = "wafer_data"
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Save the combined_data DataFrame as "wafer_data_update.csv" inside the "wafer_data" folder
file_path = os.path.join(folder_path, "wafer_data_update.csv")
combined_data.to_csv(file_path, index=False)
```
